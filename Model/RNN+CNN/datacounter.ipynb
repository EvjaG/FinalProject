```python
extOG,extNU=0,0
empty=0
def addSubFolders(path):
    '''Will return a list of all fetched files within each subfolder.'''
    folders = list(filter(lambda p: os.path.isdir(path+'/'+p),os.listdir(path)))
    files = []
    for f in folders:
        filepath = path+'/'+f
        files += addFiles(filepath) + addSubFolders(filepath)
    return files

def addFiles(path):
    '''Returns a list of all the files within a given folder '''
    files = os.listdir(path) # list of all files
    return files

def main():
    global path
    print(path)
    # exit()
    tic = time.time()    
    files = (addFiles(path))
    fileAdd = addSubFolders(path)
    files = files+fileAdd
    print(f"Counted {len(files)} files.")
    toc = time.time()
    worktime = f'Work time: {toc - tic} seconds'
    
    print(f"{worktime}\nClosing Program.")
    
    
main()
```
